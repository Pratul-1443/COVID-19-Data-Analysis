```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('covid_19.csv')
```

```python
# Display the first few rows of the dataset
df.head()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
df.columns
```

```python
# Summary statistics
summary = df.describe()
summary
```

```python
# Convert 'Confirmed', 'Deaths', and 'Recovered' to numeric (remove any commas or other formatting)
df['Confirmed'] = pd.to_numeric(df['Confirmed'], errors='coerce')
df['Deaths'] = pd.to_numeric(df['Deaths'], errors='coerce')
df['Recovered'] = pd.to_numeric(df['Recovered'], errors='coerce')
```

```python
df
```

```python
# Group by 'Country/Region' and sum the values
country_data = df.groupby('Country/Region')[['Confirmed', 'Deaths', 'Recovered']].sum().reset_index()
top_10=df.groupby('Country/Region')[['Confirmed', 'Deaths', 'Recovered']].sum().reset_index().head(10)
```

```python
country_data
```

```python
# all death, confirmed, recover data

plt.figure(figsize=(14, 8))
sns.barplot(data=country_data.melt(id_vars='Country/Region', var_name='Type', value_name='Count'), x='Country/Region', y='Count', hue='Type', palette='viridis')
plt.title('Total Confirmed Cases, Deaths, and Recoveries by Country/Region')
plt.xlabel('Country/Region')
plt.ylabel('Number of Cases')
plt.xticks(rotation=90)
plt.legend(title='Case Type')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(data=top_10.melt(id_vars='Country/Region', var_name='Type', value_name='Count'), x='Country/Region', y='Count', hue='Type', palette='viridis')
plt.title('Total Confirmed Cases, Deaths, and Recoveries by Country/Region')
plt.xlabel('Country/Region')
plt.ylabel('Number of Cases')
plt.xticks(rotation=90)
plt.legend(title='Case Type')
plt.tight_layout()
plt.show()
```

```python
colors = sns.color_palette('viridis', n_colors=len(top_10))

plt.pie(top_10["Confirmed"],labels=top_10["Country/Region"], autopct='%1.1f%%', colors=colors, startangle=140)
plt.axis('equal')

# Show plot
plt.tight_layout()
plt.show()
```
